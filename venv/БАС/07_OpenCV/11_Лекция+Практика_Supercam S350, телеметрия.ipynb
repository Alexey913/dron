** `haversine_distance_np`:**
   -     ,                 .
   -  :      (`lat1, lon1, lat2, lon2`).
   -  `R`      (6371000 ).
   -  (`lat`)   (`lon`)        `np.radians`.
   -           (`delta_phi`  `delta_lambda`).
   -    `a`   .
   -  `c`             .
   -     ,      .

**   :**
   -       DataFrame `df`,   `lat1`  `lon1`   ,  ,  `lat2`  `lon2`   ,  .             .

```python
import numpy as np

def haversine_distance_np(lat1, lon1, lat2, lon2):
    R = 6371000  #    
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)

    a = np.sin(delta_phi / 2.0)**2 + \
        np.cos(phi1) * np.cos(phi2) * \
        np.sin(delta_lambda / 2.0)**2

    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    distance = R * c
    return distance  #  

#       
lat1 = df['lat'].values[:-1]
lon1 = df['lon'].values[:-1]
lat2 = df['lat'].values[1:]
lon2 = df['lon'].values[1:]

#  
distances = haversine_distance_np(lat1, lon1, lat2, lon2)

#  
total_distance = np.sum(distances)
print(f" ,  : {total_distance:.2f} ")
```
