Параметры телеметрии
 ## Date (Дата) и Time (Время):

Это временные метки, фиксирующие точное время, когда были сделаны снимки или записаны данные. Это важно для синхронизации с другими датчиками или источниками данных, а также для последующего анализа траекторий.

 ## Latitude (lat) и Longitude (lon):

Географические координаты, определяющие местоположение объекта на поверхности Земли. В данном случае координаты указывают на конкретные точки в районе озера Ладога.
Пример использования: можно построить карту маршрута объекта на основе последовательных координат.

 ## Altitude (alt):

Высота над уровнем моря. Важно для понимания положения объекта в вертикальной плоскости, особенно при работе с беспилотными летательными аппаратами (БПЛА).

 ## Pitch (Тангаж):

Угол наклона объекта относительно горизонтальной оси. Положительное значение означает, что передняя часть объекта поднята вверх, отрицательное — что она опущена.

 ## Roll (Крен):

Угол наклона объекта относительно его продольной оси (осевая линия, проходящая через объект от носа к хвосту). Положительное значение указывает на наклон вправо, отрицательное — влево.

 ## Azimuth (Азимут):


Угол между направлением на север и направлением на объект по часовой стрелке. Азимут измеряется в градусах и показывает, в каком направлении движется объект относительно северного полюса.

 ## GPS direction (Направление GPS):

Направление движения объекта, зафиксированное с помощью данных GPS. Оно может совпадать с азимутом, но в зависимости от маневров могут возникать небольшие расхождения.

 ## GPS speed (Скорость по GPS):

Скорость объекта, измеренная с помощью GPS. Это один из ключевых показателей для оценки динамики движения.

Телеметрические данные критичны для:

Навигации и маршрутизации: Позволяют системе знать своё местоположение и строить эффективные маршруты.

Анализа полёта: Можно использовать данные для последующего анализа эффективности и безопасности полёта.

# Примеры особенностей применения телеметрии в различных условиях

## A. Воздушные системы (например, БПЛА)

Особенности:

В условиях сильного ветра необходимо контролировать углы крена и тангажа, чтобы избежать турбулентности и потери управления.

Важна высокая частота получения данных, так как изменения могут происходить быстро.

## B. Наземные системы (автономные транспортные средства)

Наземные транспортные системы используют телеметрию для навигации, предотвращения столкновений и контроля скорости.

В условиях сложного рельефа важны данные о крене и азимуте, чтобы предотвратить опрокидывание.

В городских условиях требуется более точное слежение за местоположением и низкая скорость реакции для избежания аварий.

## C. Морские системы (автономные суда)

Важно учитывать задержки передачи данных при использовании спутниковых систем в открытом море.

## D. Экстремальные условия

В экстремальных условиях, таких как полёты на больших высотах или работа в условиях сильного магнитного поля, телеметрия должна учитывать возможные искажения данных.

В высокогорных районах важно учитывать быстро меняющиеся погодные условия, которые могут влиять на данные о высоте и углах ориентации.

Этот код предназначен для вычисления общего пути, пройденного БПЛА (беспилотным летательным аппаратом), используя набор географических координат (широта и долгота) из DataFrame `df`. Он применяет формулу гаверсинуса для расчета расстояний между последовательными точками координат на поверхности Земли.

**Функция `haversine_distance_np`:**
   - Эта функция реализует алгоритм гаверсинуса, который используется для вычисления кратчайшего расстояния между двумя точками на поверхности Земли с учетом её сферической формы.
   - Входные параметры: широта и долгота двух точек (`lat1, lon1, lat2, lon2`).
   - Константа `R` представляет радиус Земли в метрах (6371000 м).
   - Широта (`lat`) и долгота (`lon`) обеих точек преобразуются в радианы с помощью `np.radians`.
   - Производится расчет разницы в радианах между соответствующими координатами двух точек (`delta_phi` и `delta_lambda`).
   - Вычисляется промежуточное значение `a` по формуле гаверсинуса.
   - Параметр `c` определяется через функцию арктангенса для нахождения центрального угла в радианах между двумя точками.
   - Возвращается итоговое расстояние в метрах, умножив центральный угол на радиус Земли.

**Создание пар последовательных точек:**
   - Извлекаются значения широты и долготы из DataFrame `df`, так что `lat1` и `lon1` содержат все точки, кроме последней, а `lat2` и `lon2` содержат все точки, кроме первой. Это делает последовательные точки из данных готовыми для передачи в функцию расчета расстояния.

```python
import numpy as np

def haversine_distance_np(lat1, lon1, lat2, lon2):
    R = 6371000  # Радиус Земли в метрах
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)

    a = np.sin(delta_phi / 2.0)**2 + \
        np.cos(phi1) * np.cos(phi2) * \
        np.sin(delta_lambda / 2.0)**2

    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    distance = R * c
    return distance  # в метрах

# Сдвигаем данные для получения пар последовательных точек
lat1 = df['lat'].values[:-1]
lon1 = df['lon'].values[:-1]
lat2 = df['lat'].values[1:]
lon2 = df['lon'].values[1:]

# Вычисляем расстояния
distances = haversine_distance_np(lat1, lon1, lat2, lon2)

# Общий путь
total_distance = np.sum(distances)
print(f"Общий путь, проделанный БПЛА: {total_distance:.2f} метров")
```
